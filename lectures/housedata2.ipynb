```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('houses.csv')
```

```python
df.head()
```

```python
# Filter a single house with certain id
single_house = df[df['id'] == 3793500160]
single_house
```

```python
# Also iloc[] works
single_house = df.iloc[13571]
single_house.to_frame()
```

```python
max_house = df.loc[df['price'].idxmax()]
max_house.to_frame()
```

<h4>Best way to do it:</h4>

```python
# nlargest() prints
df.nlargest(5, 'price')
```

```python
# 
df.nsmallest(5, 'price')
```

```python
# Average price
mean_price = df['price'].mean()
mean_price
```

```python
# Average bedroom size
mean_bedrooms = df['bedrooms'].mean()
mean_bedrooms
```

```python
# Minimum price and maximum price
min_price = df['price'].min()
max_price = df['price'].max()

# Print both values
print(min_price)
print(max_price)
```

<h4>Easiest way to do it: describe()</h4>

```python
# override default pandas scientific notation to decimal notation 
# => easier to read
pd.set_option('display.float_format', lambda x: '%.1f' % x)
```

```python
# All usual statistics easily
df.describe()
```

```python
# Check for duplicates
df.duplicated().sum()
```

```python
# filter houses price over 2M only
houses_2M = df[df['price'] > 2000000]
houses_2M
```

<h3>Group by is helpfull</h3>

```python
# Group by grade
grade_avg = df.groupby('grade').mean(numeric_only=True)
grade_avg
```

```python
# get counts by category
grade_counts = df.groupby('grade').count()
grade_counts
```

```python
# value_counts is usually best way to go
df['grade'].value_counts().to_frame()
```
