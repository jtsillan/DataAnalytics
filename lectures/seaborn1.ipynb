<h3>Seaborn and plotting</h3>

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# Load best testset from sns
# https://github.com/mwaskom/seaborn-data
tips = sns.load_dataset('tips')

# Create new column, tips is pandas DataFrame
tips['total_cost'] = tips['total_bill'] + tips['tip']

tips.head(10)
```

<b>Histogram / distribution plot - histplot()</b>

```python
# Plot total_bill
sns.histplot(tips['total_bill'])
```

```python
# Same for tip column
sns.histplot(tips['tip'])
```

```python
# Plot total cost
sns.histplot(tips['total_cost'])
```

<b>Smoother curve - KDE-line</b><br></br>
KDE = Kernel Density Estimation

```python
# Use kde=True to plot also KDE-line in histogram
sns.histplot(tips['total_bill'], kde=True)
```

```python
# Specify amount of bars with bins parameter, rarely needed
# But if normal distribution looks good, but you want extra info
sns.histplot(tips['tip'], bins=30)
```

<b>Scatter plot comparing distribution of two variables - scatterplot()</b>
<p>Used to find linear connection</p>
<p>If trend botton left to upper right --> positive correlation</p>
<p>If trend upper left to bottom right --> negative correlation</p>

```python
# Compare total_bill and tip
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

<b>Categorical plots, barplot, countplot, etc</b>

```python
# Typical barplot, show mean cost by each day
# Vertical line in each bar represents standard deviationS
sns.barplot(x='day', y='total_cost', data=tips)
```

```python
sns.barplot(x='time', y='total_cost', data=tips)
```

```python
# Custom color palette:
# https://matplotlib.org/stable/users/explain/colors/colormaps.html
sns.barplot(x='day', y='total_cost', data=tips, palette='rainbow')
```

```python
sns.barplot(x='smoker', y='total_cost', data=tips, palette='rainbow', hue='smoker')
```

```python
sns.countplot(x='smoker', data=tips)
```
