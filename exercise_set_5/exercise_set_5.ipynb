```python
# Import Beautifulsoap
from bs4 import BeautifulSoup
```

```python
# Url to get data
url = "https://en.wikipedia.org/wiki/Rovaniemi"
# Responce from get request
page = requests.get(url)
# Variable to store page content as a soap object
soap = BeautifulSoup(page.content, 'html.parser')
```

```python
# Find coordinates from soup
coord = soap.find('span', class_='geo-dms').text
# Print coordinates
coord
```

```python
# Find nicknames
# Find all elements with html element 'tr' and class 'mergedtoprow'
name_data = soap.find_all('tr', class_='mergedtoprow')
len(name_data)
```

```python
# Loop all elements
for line in name_data:
    # Find elements with data
    if line.find('td') != None:
        # Find next element with right div and class info
        info = line.findNext('div', class_='ib-settlement-nickname nickname')
        print(info.text)
        break
```

#### Extra advanced task: find population

```python
# Find all elements with html element 'tr' and class 'mergedtoprow' and 'mergedrow'
table_data = soap.find_all('tr', class_='mergedrow' and 'mergedtoprow')

# Find right element
for line in table_data:
    # Take only ones with data
    if line.find('th') != None:
        # Find previous line with only string data
        if line.find('th').find(string=True) == 'Population':
            # Find line with 'td'
            p = line.findNext('td')
            print(f"Population: {p.text} person")
```

<h4><b>6. Simple merging:</b> Download <u><b>coffee_sales.csv</b></u> and <b><u>temperatures_fixed.csv</u></b> from Moodle.<b></h4>
<p>Use merging to combine these files into one DataFrame, containing the coffee sales data and the corresponding temperature â€“data. (merge(), see materials for examples)</p>


<span style="color: red;"><b>Use left join, and coffee sales as the first DataFrame!</b></span>

```python
# Read csv file
df_coffee = pd.read_csv('coffee_sales.csv')
df_coffee
```

```python
# Read csv file
df_temp = pd.read_csv('temperatures_fixed.csv')
df_temp
```

```python
# Merge new dataframe based on Date column
df_new = df_coffee.merge(df_temp, left_on="Date", right_on="Date", how='left')
df_new
```

```python
# Extra task
# Read csv file
df_temp_unfixed = pd.read_csv('temperatures_unfixed.csv')
df_temp_unfixed
```

```python
# Drop extra columns
df_temp_unfixed = df_temp_unfixed.drop(axis=1, columns={"Summary", "Hour"})
df_temp_unfixed
```

```python
# Group by date and set mean temperature for value
df_temp_grouped = df_temp_unfixed.groupby(['Date']).mean('Temperature (C)')
df_temp_grouped
```

```python
# Merge two dataframes into one by date join left
df_merged = df_coffee.merge(df_temp_grouped, left_on='Date', right_on='Date', how='left')
df_merged
```
