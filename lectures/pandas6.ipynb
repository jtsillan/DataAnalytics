```python
import numpy as np
import pandas as pd
```

```python
# Load binary version of ranges, only two ranges values (High, Low)
df = pd.read_csv("ranges_binary.csv")
```

```python
df
```

```python
# factorize() is safe to use with binary category
# because the order doesn't matter with just two alternatives
values1, names1 = pd.factorize(df['Range'], sort=False)

# add the assigned numbers as new columns to DataFrame
df['RangeValue'] = values1
```

```python
# Since there was only two options in Range, factorize() works fine
# Can only assign values reversed (High = 0, Low = 1)
df
```

```python
# Correlation is strong, but negative due to reversed values in RangeValues
df.corr(numeric_only=True)
```
