```python
import numpy as np
import pandas as pd
```

```python
# Load csv file to DataFrame
# Dataset explanation:
# https://kaggle.com/arathipraj/house-data
df = pd.read_csv('houses.csv')
```

```python
df.head()
```

<b>Correlation matrix - housing data</b>

```python
# Correlations to separate DataFrame
# Default method is Pearson-correlation
correlations = df.corr(numeric_only=True)
```

```python
# Show correlation matrix
# Lots of columns, hard to read
correlations
```

```python
# Highlight high 10% qauntile with style
correlations.style.highlight_quantile(axis=None, q_left=0.9, color="darkgreen")
```

```python
# Highlight low 10% qauntile with style
correlations.style.highlight_quantile(axis=None, q_right=0.1, color="darkred")
```
