```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
# let's load seaborn test dataset: tips 
# these datasets are collected into a GitHub -project:
# https://github.com/mwaskom/seaborn-data
# simple datasets for basic testing with seaborn
tips = sns.load_dataset('tips')
tips['total_cost'] = tips['total_bill'] + tips['tip']
```

```python
sns.barplot(x='sex', y='tip', data=tips, hue='day')
```

```python
# Day and time regarding total cost
sns.barplot(x='day', y='total_cost', data=tips, hue='time')
```

```python
sns.barplot(x='smoker', y='total_cost', data=tips, hue='day')
```

```python
# Change palette color range
sns.barplot(x='day', y='total_cost', data=tips, hue='size', palette='bwr')
```

<b>Jointplots are handy, but can often be replaced with other plots too</b>

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
# compare total bill to tip, joint plot -> scatterplot + histogram
# use regression type, very useful and lots of information
# histogram, kde, 
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
# If you use 
sns.jointplot(x='total_bill', y='tip', data=tips, hue='time')
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips, hue='time')
```

<b>Box plots -many features from data in one plot, categorical plot</b>

```python
# this shows a box plot, which gives lots of info
# but can be hard to read at first. It shows
# deviations, min and max values quite well too.
# it also shows the number of outlying values
# that are far from the average values
sns.boxplot(x='day', y='tip', data=tips)
```

```python
sns.boxplot(x='day', y='tip', data=tips, hue='time')
```

```python
sns.boxplot(x='day', y='total_cost', data=tips, hue='size', palette='rainbow')
```

<b>Violin plots -similar to boxplot, shows density bit better</b>

```python
# Alternative to boxplot()
sns.violinplot(x='day', y='total_cost', data=tips, hue='time')
```

```python
sns.violinplot(x='day', y='tip', data=tips, hue='smoker', split=True)
```

```python
# Spripplot is one option to boxplot, quite easy to read
sns.stripplot(x='day', y='total_bill', data=tips, hue='smoker', jitter=True, dodge=True)
```

<b>Regression plots are useful in finding hidden connections beyond basic correlations</b>

```python
sns.lmplot(x='total_bill', y='tip', data=tips)
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='time')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='day')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='smoker')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='time')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='time', row='sex')
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', hue='sex', aspect=0.6)
```
