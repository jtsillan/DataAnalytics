```python
# Import Beautifulsoap
from bs4 import BeautifulSoup
```

```python
# Url to get data
url = "https://en.wikipedia.org/wiki/Rovaniemi"
# Responce from get request
page = requests.get(url)
# Variable to store page content as a soap object
soap = BeautifulSoup(page.content, 'html.parser')
```

```python
# Find coordinates from soup
coord = soap.find('span', id='coordinates').text
# Some parsing would be nice, but it's good enough.
coord
```

```python
# Find all elements with html element 'tr' and class 'mergedtoprow'
nicknames = soap.find_all('tr', class_='mergedtoprow')
len(nicknames)
```

```python
for col in nicknames:
    #print(col)
    if col.find('td') != None:        
        #print(col.text)
        info = col.findNext('div')
        print(col)
        print(info)
        print('----------------------------')
        print(info.text)
        print('****************************')

```
